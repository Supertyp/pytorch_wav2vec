```python
import xml.etree.ElementTree as ET


# Function to parse TIME_ORDER tag
def parse_time_order(root):
    time_slots_dict = {}
    for time_slot_elem in root.iter('TIME_SLOT'):
        
        time_slot_id = time_slot_elem.attrib.get('TIME_SLOT_ID')
        time_value = time_slot_elem.attrib.get('TIME_VALUE')

        if time_slot_id is not None and time_value is not None:
            time_slots_dict[time_slot_id] = time_value
            
    return time_slots_dict

# Function to parse annotation elements
def parse_annotations(root, time_slots_dict):
    
    annotations_dict = {}

    # Assuming 'TIER' is the parent tag of 'ANNOTATION' elements
    for tier_elem in root.iter('TIER'):
        print (tier_elem.attrib)
        for annotation_elem in tier_elem.iter('ANNOTATION'):
            #print(annotation_elem.attrib)
            for alignable_annotation_elem in annotation_elem.iter('ALIGNABLE_ANNOTATION'):
                annotation_id_elem = alignable_annotation_elem.attrib['ANNOTATION_ID']
                print (alignable_annotation_elem.attrib)

                
                time_slot_ref1_elem = alignable_annotation_elem.attrib['TIME_SLOT_REF1']
                time_value1 = time_slots_dict.get(time_slot_ref1_elem)
        
                time_slot_ref2_elem = alignable_annotation_elem.attrib['TIME_SLOT_REF2']
                time_value2 = time_slots_dict.get(time_slot_ref2_elem)
                annotation_value_List = alignable_annotation_elem.iter('ANNOTATION_VALUE')
                
                textString = ""
                for each in annotation_value_List:
                    if each.text:
                        textString = textString + each.text
                #print( annotation_id_elem, textString, time_value1, time_value2)


def parse_xml(xml_file_path):
    try:
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        time_slots_dict = parse_time_order(root)

        annotation_dict = parse_annotations(root, time_slots_dict)

        return time_slots_dict, annotations_dict

    except Exception as e:
        
        print(f"Error parsing XML file: {e}")
        return None, None

xml_file_path  = 'C:\\Users\\barth\\gits\\pytorch_wav2vec\\test_data\\Bcnt_AEF_032_Camila.eaf'

time_slots_dict, annotations_dict = parse_xml(xml_file_path)
```

```python
from pympi import Eaf
import pydub
from pydub import AudioSegment

def split_audio_by_annotations(audio_path, elan_path, output_folder):
    # Load ELAN file
    eaf = Eaf(elan_path)

    # Load audio file
    audio = AudioSegment.from_file(audio_path)

    for time_slot in eaf.time_order:
        print 

    
    # Iterate through annotations and split audio
    for tier in eaf.tiers:
        for annotation in eaf.tiers[tier]:
            for k, v in annotation.items():
            
                
                start_time = v[0]
                print(start_time)
                
                end_time = v[1]
                annotation_text = v[2]
    
                # Extract the corresponding audio segment
                audio_segment = audio[start_time:end_time]
    
                # Output file path
                output_path = f"{output_folder}/{annotation_text}.wav"
    
                # Save the audio segment to a file
                audio_segment.export(output_path, format="wav")
    
                print(f"Segment '{annotation_text}' saved to {output_path}")

# Example usage
audio_file_path = 'C:\\Users\\barth\\gits\\pytorch_wav2vec\\test_data\\Bcnt_AEF_032_Camila.wav'
elan_file_path = 'C:\\Users\\barth\\gits\\pytorch_wav2vec\\test_data\\Bcnt_AEF_032_Camila.eaf'
output_folder_path = 'C:\\Users\\barth\\gits\\pytorch_wav2vec\\test_data\\output'

split_audio_by_annotations(audio_file_path, elan_file_path, output_folder_path)
```
